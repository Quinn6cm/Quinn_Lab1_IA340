## Visualize Data

```python
import matplotlib.pyplot as plt #This allows me to add labels and edit the visualizations that I create
```

### Films By Year Visual

```python
df_filtered = df[df["Release Year"] >= 1980] 
#This is used to chart filtered data as there are almost no movies in the data set before 1980
films_by_year = df_filtered.groupby("Release Year")["Title"].nunique() # Count unique films per year
films_by_year.tail()
```

```python
films_by_year.plot() #This plots the data for films per year after 1980.
plt.title("Unique Films by Release Year Filmed in San Francisco (After 1980)")
plt.xlabel("Release Year")
plt.ylabel("Number of Unique Films Filmed in San Francisco")
plt.show()
```

The above data shows how many films had filming locations in San Fransico from 1980-2023. One trend I noticed is there was a giant spike around 2015-2016 which included almost double the amount of movies filmed in SF. The normal amount was around 6-8 with a few exceptions during the years.

### Film Locations In SF Visual

```python
# Look at min/max of coordinates
print(df["Longitude"].min(), df["Longitude"].max())
print(df["Latitude"].min(), df["Latitude"].max())
# Filter out rows with missing coordinates
df_map = df.dropna(subset=["Longitude", "Latitude"])
#My first attempt at making the scatter plot showed one location at -121 which was not located in San Fran so I had to filter the data a bit more
```

```python
df_map = df[
    (df["Longitude"].between(-123, -122)) &  # valid SF longitudes
    (df["Latitude"].between(37, 38))        # valid SF latitudes
]
#This will stop the data from going too far outside of San Fran
```

```python
plt.scatter(
    df_map["Longitude"],
    df_map["Latitude"],   
    alpha=0.6,  # transparency of points (0 = invisible, 1 = solid)
    s=40,       # size of each point
    c="steelblue"   # color of points
)
plt.title("Film Locations in San Francisco")  
plt.xlabel("Longitude")                        
plt.ylabel("Latitude")                        
```

The above scatterplot displays the latitude and longitude of film locations in SF. There was one location in the data outside of SF which I had to filter out. The majority of locations seems to be in central SF or the others are in the outskirts. If we had a map we could further analyze the locations of filming.
